### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
a = np.random.randint(0, 10, (6, 4))
b = np.random.randint(0, 20, (4, 3))
print(a)
print(b)
c = np.dot(a, b)
print(c)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
from scipy import stats
a = np.random.rand(7, 7)
print(f"Исходный массив: \n{a}\n")
a_mean = np.mean(a, axis=0)
a_std = np.std(a, axis=0)
print(f"mean: {a_mean} \n\nstd: {a_std}\n")
a_norm = (a - a_mean) / a_std
print(f"Стандартизированный массив: \n{a_norm}\n")
a_norm_scipy = stats.zscore(a)
print(f"Стандартизированный массив через SciPy: \n{a_norm}")
```
